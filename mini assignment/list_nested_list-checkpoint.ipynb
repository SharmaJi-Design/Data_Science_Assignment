### make the list and send to the other list and make it nested list

```python
list1 = [10, 20, 30]
list2 = []

list2.append(list1)

first_value = list1[0]
list2.append(first_value)

removed_item = list2.pop()

print("List1:", list1)
print("List2 (nested):", list2)
print("Removed item:", removed_item)
```
